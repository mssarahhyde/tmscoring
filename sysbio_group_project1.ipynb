```python
import subprocess

gse_accessions = list(df['GEO accession'].values)

for accession in gse_accessions:
  gse = GEOparse.get_GEO(geo=accession, destdir="./")
  for gsm in gse.gsms:
    if gse.gsms[gsm].table is not None and 'VALUE' in gse.gsms[gsm].table:
      exp_matrix = pd.concat([exp_matrix, gse.gsms[gsm].table['VALUE']], axis=1)
      exp_matrix = exp_matrix.rename(columns={'VALUE': gsm})
  subprocess.run(['rm', f'{accession}_family.soft.gz'])
```

```python
exp_matrix
```

```python
exp_matrix.to_csv('yeast_wgcna_expression_data_raw.csv')
```
